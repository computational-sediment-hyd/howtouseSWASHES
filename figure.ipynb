```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### case1

```python
fpath = "sol11.dat"
df = pd.read_csv(fpath, sep=r"\s+", skiprows=18)
print(df.columns)
df.columns = ["L", "h", "u", "zb", "q", "zb+h", "Fr", "zb+hc"]
print(df.columns)
```

```python
fpath = "sol12.dat"
df2 = pd.read_csv(fpath, sep=r"\s+", skiprows=18)
print(df2.columns)
df2.columns = ["L", "h", "u", "zb", "q", "zb+h", "Fr", "zb+hc"]
print(df2.columns)
```

```python
%matplotlib inline

sns.set(context='paper', palette="bright", style="whitegrid" , font_scale=1)

fig = plt.figure(dpi=150, figsize=(5, 4))
grid = plt.GridSpec(2, 1, wspace=0.1, hspace=0.1, height_ratios=[1, 1])
ax1 = fig.add_subplot(grid[0, 0])
ax2 = fig.add_subplot(grid[1, 0], sharex=ax1)


ax1.plot([0,1000,1000,2000], [6,6,0,0], label="Initial", color="k", lw=4)
ax1.plot(df["L"], df["zb+h"], label="nx=20", marker="o", markersize=5)
ax1.plot(df2["L"], df2["zb+h"], label="nx=200", linestyle="dashed", marker="o", markersize=1)

ax2.plot(df["L"], df["u"], label="nx=20", marker="o", markersize=5)
ax2.plot(df2["L"], df2["u"], label="nx=200", linestyle="dashed", marker="o", markersize=1)

ax1.set_xlim(0, 2000)

# grid
for axp in [ax1,ax2]:
    axp.minorticks_on()
    axp.grid(which = "minor", axis="both", lw=0.3, ls='-', color="lightgray")

# labels
ax1.tick_params(labelbottom=False)
ax1.set_ylabel("waterlevel (m)")
ax2.set_xlabel("distance (m)")
ax2.set_ylabel("velocity (m/s)")

ax1.legend(loc="upper right")
ax1.set_title("Case1( t=40sec )")

fig.savefig(r'calcoutC1.png', bbox_inches='tight', dpi=150)

plt.show()
```

### case2

```python
fpath = "sol22.dat"
df = pd.read_csv(fpath, sep=r"\s+", skiprows=27, skipfooter=1, engine="python")
print(df.columns)
df.columns = ["L", "h", "u", "zb", "q", "zb+h", "Fr", "zb+hc","zbini","zbini+h"]
print(df.columns)
```

```python
%matplotlib inline

sns.set(context='paper', palette="bright", style="whitegrid" , font_scale=1)

fig = plt.figure(dpi=150, figsize=(5, 3))
grid = plt.GridSpec(1, 1, wspace=0.1, hspace=0.1)#, height_ratios=[1, 1])
ax1 = fig.add_subplot(grid[0, 0])
# ax2 = fig.add_subplot(grid[1, 0], sharex=ax1)

ax1.plot(df["L"], df["zb+h"], label="H", c="b")
ax1.plot(df["L"], df["zbini+h"], label="H_initial", c="b", linestyle="dashed")
ax1.plot(df["L"], df["zb"], label="zb", c="k")
ax1.plot(df["L"], df["zbini"], label="zb_initial", c="k", linestyle="dashed")

ax1.set_xlim(0, 15)

# grid
for axp in [ax1,ax2]:
    axp.minorticks_on()
    axp.grid(which = "minor", axis="both", lw=0.3, ls='-', color="lightgray")

# labels
ax1.set_ylabel("waterlevel (m)")
ax1.set_xlabel("distance (m)")

ax1.legend(loc="upper right")
ax1.set_title("Case2( t=7sec )")

fig.savefig(r'calcoutC2.png', bbox_inches='tight', dpi=150)

plt.show()
```
